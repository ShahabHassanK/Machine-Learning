```python
from math import sqrt
import numpy as np
import warnings
import matplotlib.pyplot as plt
from matplotlib import style
from collections import Counter
style.use('fivethirtyeight')

dataset = {'a': [[2,2],[3,5],[4,6]], 'b': [[3,6], [4,7], [5,6]]}
new_point = [2,2]
color_map = {'a':'b','b':'g','c':'r'}
for i in dataset:
    for ii in dataset[i]:
        plt.scatter(ii[0], ii[1], s=100, color = color_map[i])
plt.scatter(new_point[0], new_point[1], s=100, color='y')
plt.show()
```

```python
def KNN(data, predict, k=5):
    if len(data) >= k:
        warnings.warn('K is set to a value smaller than total number of voting groups')
    distances = []
    for group in data:
        for features in data[group]:
            euclidean_distance = np.linalg.norm(np.array(features) - np.array(predict))
            distances.append([euclidean_distance, group])
    votes = [i[1] for i in sorted(distances) [:k]]
    print('Vote details: ', Counter(votes).most_common(1))
    vote_results = Counter(votes).most_common(1)[0][0]
    return vote_results
result = KNN(dataset, new_point, k=5)
print(result)
```
