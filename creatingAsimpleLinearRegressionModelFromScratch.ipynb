```python
from statistics import mean
import numpy as np
import matplotlib.pyplot as plt
import math
from matplotlib import style
#Testing Assumptions
import random
```

```python
# x = np.array([1,2,3,4,5,6], dtype=np.float64)
# y = np.array([5,4,6,5,6,7], dtype=np.float64)                                                            
  
  
#Creating sample dataset containing random values for testing the assumptions  

def create_dataset(n, variance, step=2, correlation=False):
    val = 5
    y = []
    for i in range(n):
        y_val = val + random.randrange(-variance, variance)
        y.append(y_val)
        if correlation and correlation == 'pos':
            val += step
        elif correlation and correlation == 'neg':
            val -+ step
    x = [i for i in range(len(y))]    
    return np.array(x, dtype=np.float64), np.array(y, dtype=np.float64)
    
x, y = create_dataset(100, 20, 4, correlation = 'pos')                    
                                           
def best_fit_slope(x,y):
    m = ( ((mean(x) * mean(y)) - mean(x*y)) 
        / ((mean(x) ** 2) - (mean(x**2))) )
    return m
m = best_fit_slope(x,y)

def intercept(x,y):
    b = mean(y) - (m * mean(x))
    return b
b = intercept(x,y)    
print("slope = ", m, 'intercept = ', b)
regression_line = [(m*x_val) + b for x_val in x]                                                                                                                                          
```
