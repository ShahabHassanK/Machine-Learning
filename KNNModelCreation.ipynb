```python
from math import sqrt
import numpy as np
import warnings
import matplotlib.pyplot as plt
from matplotlib import style
from collections import Counter
style.use('fivethirtyeight')
#SIMPLE WAY TO GET EUCLIDEAN DISTANCE
#p1 = [1, 3]
#p2 = [2, 5]
#euclidean_distance = sqrt( (p1[0]-p2[0])**2 + (p1[1] - p2[1])**2 )
#euclidean_distance


dataset = {'a': [ [1,2], [2,3], [3,1] ], 'b' : [ [6,5], [7,7], [8,6] ]}
new_feature = [5, 7]
color_map = {'a': 'r', 'b': 'g'}

for i in dataset:
    for ii in dataset[i]:
        plt.scatter(ii[0], ii[1], s=100, color= color_map[i])
plt.scatter(new_feature[0], new_feature[1], s = 100, color = 'k')
plt.show()

def KNN(data, predict, k=5):
    if len(data) >= k:
        warnings.warn('K is set to a value less than total voting groups..')
    #knnalgos
    distances = []
    for group in data:
        for features in data[group]:
            #Numpy method to calculate euclidean distance (faster and usable for more data points)
            euclidean_distance = np.sqrt((np.sum(np.array(features) - np.array(predict))**2))
            #even faster one below
            #euclidean_distance = np.linalg.norm(np.array(features) - np.array(predict))
            distances.append([euclidean_distance, group])
    votes = [i[1] for i in sorted(distances) [:k]]
    print(Counter(votes).most_common(1))
    vote_results = Counter(votes).most_common(1)[0][0]
    return vote_results

result = KNN(dataset, new_feature, k=5)
print(result)
           
```
