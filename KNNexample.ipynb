```python
import pandas as pd
import numpy as np
from sklearn import preprocessing, model_selection, neighbors
import pickle
```

```python
df = pd.read_csv('breast-cancer-wisconsin.data')
df.replace('?', -99999, inplace=True)
df.drop(['id'], axis= 1,inplace=True)
df
#In the dataset in class column there are two classes '2' & '4'. 2 represents benign and 4 represents malignant
```

```python
#Defining features and label
#label = class and all other columns are features
x = np.array(df.drop(['class'], axis=1))
y = np.array(df['class'])
x_train, x_test, y_train, y_test = model_selection.train_test_split(x, y, test_size= 0.2)
clf = neighbors.KNeighborsClassifier()
clf.fit(x_train, y_train)                                                                                                                                                                                    
  
  
#PICKLING IF NECESSARY< NOT NECESSARY HERE
# with open('KNN1.pickle', 'wb') as file:
#     pickle.dump(clf,file)
pickle_in = open('KNN1.pickle', 'rb') 

clf = pickle.load(pickle_in)
accuracy = clf.score(x_test, y_test)
accuracy
```

```python
#creating a random data row to testing whether it belongs to class '2' or '4'
example_measure = np.array([[4,2,1,1,1,2,3,2,1], [4,3,1,1,2,2,3,2,1], [4,1,2,2,3,2,3,2,2]])
example_measure = example_measure.reshape(len(example_measure),-1)
prediction = clf.predict(example_measure)
print(prediction, accuracy)
#SO the predicted class for the given example measure is 2 for all 3 inputs.
```
